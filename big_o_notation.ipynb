```python
import matplotlib.pyplot as plt
```

```python
import math
```

```python
plt.figure(figsize = (16,9))

# first plot
# y = 0x + 1 and label the curve "O(1)"
x1_axis=[]
y1_axis=[]
for x in range(-5, 6):  
    y = (0)*x + 1
    x1_axis.append(x)
    y1_axis.append(y)    
plt.plot(x1_axis,y1_axis, color = 'red', label = 'O(1)')

# second plot
# y = log(x) and label the curve "O(log n)"
x2_axis=[]
y2_axis=[]
for x in range(1, 6):  
    y = math.log(x)
    x2_axis.append(x)
    y2_axis.append(y)    
plt.plot(x2_axis,y2_axis, color = 'orange', label = 'O(log n)')

# third plot
# y = x and label the curve "O(n)"
x3_axis=[]
y3_axis=[]
for x in range(-5, 6):  
    y = x
    x3_axis.append(x)
    y3_axis.append(y)    
plt.plot(x3_axis,y3_axis, color = 'yellow', label = 'O(n)')

# fourth plot
# y = x ∗ log(x) and label it "O(n log n)"
x4_axis=[]
y4_axis=[]
for x in range(1, 6):  
    (x) * math.log(x)
    x4_axis.append(x)
    y4_axis.append(y)    
plt.plot(x4_axis,y4_axis, color = 'limegreen', label = 'O(n log n)')
         


# fifth plot
# y = x^2 and label it "O(n^2)"
x5_axis=[]
y5_axis=[]
for x in range(-1, 6):  
    y = x^2
    x5_axis.append(x)
    y5_axis.append(y)    
plt.plot(x5_axis,y5_axis, color = 'green', label = 'O(n^2)')

# sixth plot
# y = 2^x and label it "O(2^n)"
x6_axis=[]
y6_axis=[]
for x in range(-5, 1):  
    y = 2^x
    x6_axis.append(x)
    y6_axis.append(y)    
plt.plot(x6_axis,y6_axis, color = 'deepskyblue', label = 'O(2^n)')

# seventh plot
# y = n! and label it "O(n!)"
x7_axis=[]
y7_axis=[]
for n in range(0, 5):  
    y = math.factorial(n)
    x7_axis.append(x)
    y7_axis.append(y)    
plt.plot(x7_axis,y7_axis, color = 'royalblue', label = 'O(n!)')

# eighth plot
# y = n^n and label it "O(n^n)"
x8_axis=[]
y8_axis=[]
for x in range(-5, 6):  
    y = n^n
    x8_axis.append(x)
    y8_axis.append(y)    
plt.plot(x8_axis,y8_axis, color = 'purple', label = 'O(n^n)')


plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='upper left')
plt.grid(True)
plt.show() 
```
