```python
#1. Use matplotlib to plot the following equation:

# y = x^2 − x + 2

#You'll need to write the code that generates the x and y points.
#Add an anotation for the point 0, 0, the origin.
```

```python
import matplotlib.pyplot as plt
```

```python
# y = x^2 − x + 2

x_axis=[]
y_axis=[]

for x in range(-50, 50):
    y= x**2 - x +2
    x_axis.append(x)
    y_axis.append(y)

plt.plot(x_axis,y_axis)
plt.grid(True)
plt.show()
```

```python
#2. Create and label 3 separate charts for the following equations (choose a range for x that makes sense):

# y = √ x
# y = x^3
# y = 2^x

#You can use functions from the math module to help implement some of the equations above.
```

```python
import math
```

```python
# y = √ x

x_axis=[]
y_axis=[]

for x in range(1, 50):
    y = math.sqrt(x)
    x_axis.append(x)
    y_axis.append(y)

plt.plot(x_axis,y_axis)
plt.grid(True)
plt.show()
```

```python
# y = x^3

x_axis=[]
y_axis=[]

for x in range(-50, 50):
    y = x**3
    x_axis.append(x)
    y_axis.append(y)

plt.plot(x_axis,y_axis)
plt.grid(True)
plt.show()
```

```python
# y = 2^x

x_axis=[]
y_axis=[]

for x in range(-15, 15):
    y= y = 2**x
    x_axis.append(x)
    y_axis.append(y)

plt.plot(x_axis,y_axis)
plt.grid(True)
plt.show()
```

```python
#3. Combine the figures you created in the last step into one large figure with 3 subplots.

n_rows = 1
n_cols = 4

# first subplot
# y = x^2 − x + 2
x1_axis=[]
y1_axis=[]
for x in range(-10, 10):
    y= x**2 - x + 2
    x1_axis.append(x)
    y1_axis.append(y)
plt.subplot(2,2,1)
plt.plot(x1_axis,y1_axis, color = 'red')
plt.title('y = x^2 − x + 2')
plt.grid(True)


# second subplot
# y = √ x
x2_axis=[]
y2_axis=[]
for x in range(1, 10):
    y = math.sqrt(x)
    x2_axis.append(x)
    y2_axis.append(y)
plt.subplot(2,2,2)
plt.plot(x2_axis,y2_axis, color = 'yellow')
plt.title('y = √ x')
plt.grid(True)


# third subplot
# y = x^3
x3_axis=[]
y3_axis=[]
for x in range(-10, 10):
    y = x**3
    x3_axis.append(x)
    y3_axis.append(y)
plt.subplot(2,2,3)
plt.plot(x3_axis,y3_axis, color = 'limegreen')
plt.title('y = x^3')
plt.grid(True)


# fourth subplot
# y = 2^x
x4_axis=[]
y4_axis=[]
for x in range(-10, 10):
    y= y = 2**x
    x4_axis.append(x)
    y4_axis.append(y)
plt.subplot(2,2,4)
plt.plot(x4_axis,y4_axis, color = 'royalblue')
plt.title('y = 2^x')
plt.grid(True)

plt.tight_layout()
plt.show()
```
